```python
import pandas as pd
import time

def bubblesort(list1):  
    has_swapped = True  
    total_iteration = 0  
    while(has_swapped):  
        has_swapped = False  
        for i in range(len(list1) - total_iteration - 1):  
            if list1[i] > list1[i+1]:  
                list1[i], list1[i+1] = list1[i+1], list1[i]  
                has_swapped = True  
        total_iteration += 1  

def insertionSort(arr):
    n=len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i-1
        while j >=0 and key < arr[j] :
                arr[j+1] = arr[j]
                j -= 1
        arr[j+1] = key
```

```python
# Measure execution time for Bubble Sort
algorithm_time = pd.DataFrame({'10^x':[],
                                'insertion':[],
                                'bubble':[]})

algorithm_time=pd.DataFrame(algorithm_time.groupby("10^x").sum())

def timing_algorithms(x:int):
    large_list = list(range(x, 0, -1))
    start_time = time.time()
    bubblesort(large_list)
    end_time = time.time()
    bubble_sort_time = end_time - start_time
    large_list = list(range(x, 0, -1))
    start_time = time.time()
    insertionSort(large_list)
    end_time = time.time()
    insertion_sort_time = end_time - start_time
    algorithm_time.loc[x] = [ insertion_sort_time*1000, bubble_sort_time*1000] 
#     print(x)
#     print("Bubble Sort execution time:", bubble_sort_time*1000)
#     print("Insertion Sort execution time:", insertion_sort_time*1000)
```

```python

fives=[5*10**(i+1) for i in range(4)]
fifi=[1.5*10**(i+1) for i in range(4)]
t10=[10**(i+1) for i in range(4)]
nums=[]
for i in(list(zip(t10,fifi,fives))):
    nums=nums+list(i)
del nums[-1]
nums=nums+[j*1000 for j in range(2,5)]+[x*1000 for x in range(6,10)]+[l*1000 for l in range(11,15)]
nums.sort()
print(nums)   
```

```python
for val in nums: 
    val=int(val)
    timing_algorithms(val)
```
